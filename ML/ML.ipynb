```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
```

```python
dataset=pd.read_csv('titanic_data (1).csv')
dataset.head(5)
```

```python
dataset.isnull().sum()
```

```python
dataset[dataset.duplicated()]
```

```python
dataset=dataset.drop_duplicates()
```

```python
dataset.duplicated().sum()
```

```python
dataset = dataset.drop('Cabin' , axis=1)
```

```python
print(dataset.isna().sum())


dataset=dataset.dropna(inplace=False)  

print(dataset.isna().sum())
```

```python
dataset= dataset.drop('Name' , axis=1)
dataset= dataset.drop('PassengerId' , axis=1)
dataset= dataset.drop('Ticket' , axis=1)
l = LabelEncoder()
dataset['Sex'] = l.fit_transform(dataset['Sex'])
dataset['Embarked'] = l.fit_transform(dataset['Embarked'])
dataset
```

```python
from sklearn.preprocessing import MinMaxScaler

# Sample dataset (replace with your actual data)

dataset = pd.DataFrame(dataset)

# Create the MinMaxScaler
scaler = MinMaxScaler(feature_range=(0, 5))

# Fit and transform the dataset
dataset = scaler.fit_transform(dataset)



dataset = pd.DataFrame(dataset, columns=['Survived','Pclass','Sex','Age','SibSp','Parch','Fare','Embarked'])

# Now you can use .head() on X_train_scaled_df
dataset.head(3)
```

```python
dataset1=dataset
dataset2=dataset
```

```python
from sklearn.model_selection import train_test_split
y = dataset1['Survived']
X = dataset1.drop(columns=['Survived'])
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
from sklearn.linear_model import Lasso

linlasso = Lasso(alpha=0.001)
linlasso.fit(X_train, y_train)

print('lasso regression linear model intercept: {}'
     .format(linlasso.intercept_))
print('lasso regression linear model coeff:\n{}'
     .format(linlasso.coef_))
print('Non-zero features: {}'
     .format(np.sum(linlasso.coef_ != 0)))
print('R-squared score (training): {:.3f}'
     .format(linlasso.score(X_train, y_train)))
print('R-squared score (test): {:.3f}\n'
     .format(linlasso.score(X_test, y_test)))
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score,classification_report, confusion_matrix

clf=DecisionTreeClassifier()
clf.fit(X_train, y_train)

# Predict on the test set
y_pred = clf.predict(X_test)

# Evaluate the model's performance
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
print("Classification Report:\n", classification_report(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

```python
y1 = dataset2['Survived']
X1 = dataset2.drop(columns=['Survived'])
X_train1, X_test1, y_train1, y_test1 = train_test_split(X1, y1, test_size = 0.2)
```

```python
from sklearn.linear_model import Ridge

linridge = Ridge(alpha=0.2)
linridge.fit(X_train1, y_train1)

print('ridge regression linear model intercept: {}'
     .format(linridge.intercept_))
print('ridge regression linear model coeff:\n{}'
     .format(linridge.coef_))
print('R-squared score (training): {:.3f}'
     .format(linridge.score(X_train, y_train)))
print('R-squared score (test): {:.3f}'
     .format(linridge.score(X_test, y_test)))
print('Number of non-zero features: {}'
     .format(np.sum(linridge.coef_ != 0)))
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score,classification_report, confusion_matrix

clf1=DecisionTreeClassifier()
clf1.fit(X_train1, y_train1)

# Predict on the test set
y_pred1 = clf1.predict(X_test1)

# Evaluate the model's performance
accuracy1 = accuracy_score(y_test1, y_pred1)
print("Accuracy:", accuracy1)
print("Classification Report:\n", classification_report(y_test1, y_pred1))
print("Confusion Matrix:\n", confusion_matrix(y_test1, y_pred1))
```
